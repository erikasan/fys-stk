# Exercise 1: Our first Python encounter

```python
# import packages
import pandas as pd
from IPython.display import display
import numpy as np

# read in data
url = "MachineLearningMSU-FRIB2020/doc/pub/Regression/ipynb/datafiles/EoS.csv"
data = pd.read_csv(url, names=('Density', 'Energy'))
data_pandas = pd.DataFrame(data)

# extract density column
density = data_pandas['Density']

# set up design matrix
X = np.zeros((len(density),5))
X[:,0] = 1
X[:,1] = density**(2.0/3.0)
X[:,2] = density**(3.0/3.0)
X[:,3] = density**(4.0/3.0)
X[:,4] = density**(5.0/3.0)

# display matrix
dmatrix = pd.DataFrame(X)
dmatrix.columns = ['1', 'density^(2/3)', 'density', 'density^(4/3)', 'density^(5/3)']
display(dmatrix)
```

# Exercise 2: making your own data and exploring scikit-learn

```python
# import packages
import pandas as pd
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt

# define x and y values with 100 data points
x = np.random.rand(100,1)
y = 2.0+5*x*x+0.1*np.random.randn(100,1)


# set up design matrix
X = np.zeros((len(x),3))
X[:,0] = 1.0
X[:,[1]] = x
X[:,[2]] = x**2

# matrix inversion to find beta
beta = np.linalg.inv(X.T @ X) @ X.T @ y


# plot fit
def f(x):
    f_x = beta[2]*x**2 + beta[1]*x + beta[0]
    return f_x
plt.plot(x,f(x))
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# aother way to plot fit
ytilde = X @ beta
plt.plot(x, ytilde)
plt.show()
```

### With scikit-learn

```python
import pandas as pd
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

def R2(y_data, y_model):
    return 1 - np.sum((y_data - y_model) ** 2) / np.sum((y_data - np.mean(y_data)) ** 2)

def MSE(y_data,y_model):
    n = np.size(y_model)
    return (1/n) * np.sum((y_data - y_model)**2)
    

# We split the data in test and training data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# matrix inversion to find beta
beta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train

# and then make the prediction
ytilde2 = X_train @ beta
plt.plot(x, ytilde2)
plt.show()

print("Training R2")
print(R2(y_train,ytilde2))
print("Training MSE")
print(MSE(y_train,ytilde2))
ypredict = X_test @ beta
print("Test R2")
print(R2(y_test,ypredict))
print("Test MSE")
print(MSE(y_test,ypredict))
```

* R2 and MSE both look good. R2 is close to 1 and MSE is close to zero which suggests a good model
